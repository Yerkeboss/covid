```python
import pandas as pd
from sklearn.cluster import KMeans


data = pd.read_csv('severe_cases_score_data.csv')

data.head()
```

```python
data.columns
```

```python

df = data.drop(columns=['Unnamed: 0', 'FIPS', 'State', 'County','Severe COVID Case Complications'], axis=1)

df
```

```python
for column in df.columns:
    df[column] = df[column] / df[column].max()
```

```python
df
```

```python
inertia = []
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=42)
    kmeans.fit(df)
    inertia.append(kmeans.inertia_)
```

```python
import matplotlib.pyplot as plt
plt.plot(range(1, 11), inertia)
plt.title('Elbow Method')
plt.xlabel('Number of Clusters')
plt.ylabel('Inertia')
plt.show()
```

```python
kmeans = KMeans(n_clusters=3, init='k-means++', max_iter=300, n_init=10, random_state=0)
df['cluster'] = kmeans.fit_predict(df)
```

```python
centroids = pd.DataFrame(kmeans.cluster_centers_, columns=df.columns[:-1])
```

```python
centroids
```

```python

# Calculate the distance between each area and its cluster centroid
df['distance'] = df.apply(lambda row: ((row[:-2] - centroids.loc[row['cluster']]) ** 2).sum(), axis=1)

# Calculate Covid19 vulnerability index for each area
df['vulnerability_index'] = df.apply(lambda row: row['distance'] / (row['distance'] + 1), axis=1)
```

```python
df.sort_values(by='vulnerability_index', ascending=False)
```

```python
X = df.drop(columns=['vulnerability_index'])
y = df[['vulnerability_index']]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42, test_size=0.2)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression().fit(X_train, y_train)
```

```python
y_pred = lr.predict(X_test)
#y_pred
```

```python
from sklearn.metrics import r2_score, mean_squared_error
r2_score(y_test, y_pred)
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

def evaluate_regression_model(y_test, y_pred):

    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    r2 = r2_score(y_test, y_pred)

    return mae, mse, r2
```

```python
mae, mse, r2 = evaluate_regression_model(y_test, y_pred)
```

```python
mae, mse, r2
```
